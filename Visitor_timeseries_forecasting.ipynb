# Visitor Time series forecasting

## Background

While often negleted, time series forecasting is a very important area of machine learning. There are many prediction problems thjat involve a time component. It is important to be aware of this time component when making predictions as it does make time series more difficult to handle by using simple prediction methods. 

## Project Description

In this project, I will use “Visitors.csv” data to forecast 24 months of monthly number of
visitors to a country following the last period in the dataset. The aim of the project is to develop univariate forecasting models using only the historical number of visitors.

In this project, I decided to explore the use of a few different timeseries forecasting methods and compare the different model performances. 
- SARIMA (Seasonal Autoregressive Integrated Moving Average)
- Holt Winters (Multiplicative)
- Simple Random Walk
- Simple Exponential Smoothing
- Holt-Winters Multiplicative Damping
- Holt Winters (Additive)

## Steps to take to implement (S)ARIMA

4) Obtain SARIMA Parameters <br />

### 1) Check For Seasonality
##### Simple Plot



By observing Seasonal, we can observen that the data is seasonal as we observe a similar trend whenever the data is of a certain season yearly. As such, we would not be able to use **ARIMA** to forecast as it does not support Seasonal Data (Time series data with a trend) so for forecasting, **SARIMA** will be used. <br />
Additionally, since there is Seasonality, the model used should be **multiplicative**

### 2) Check for data Stationarity
Will use this function to test for stationarity by using the **Augmented Dickey-Fuller test** <br />
where <br />
$H_0$: If failed to be rejected, time series data is not stationary <br />
$H_1$: If $H_0$ is rejected, it means the time series data is stationary

```python
def test_stationarity(timeseries):
    
    #Determine rolling statistics
    movingAverage = timeseries.rolling(window=12).mean()
    movingSTD = timeseries.rolling(window=12).std()
    
    #Plot rolling statistics
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
    std = plt.plot(movingSTD, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.xlabel("Date")
    plt.ylabel("Number of Visitors (In Millions)")
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    
    #Perform Dickey–Fuller test:
    print('Results of Dickey Fuller Test:')
    dftest = adfuller(timeseries['Number of Visitors'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    
```

```python
test_stationarity(db)
```

For this, we will selected the Critical value of 5% to represent $\alpha$ = -2.871388 <br />
The ADF Test Statistic observed is 0.670592 <br />
As ADF Test Statistic > $\alpha$, We fail to reject $H_0$ and the data is not stationary <br />
Another method to check, would be to refer to the p-value and compare it with the $\alpha$ of 0.05. If $\alpha$ > p-value, the data is stationary, otherwise the data is not stationary. <br />
As $\alpha$ < p-value, we can conclude that the data is not stationary

## Making Data Stationary
As the data is not stationary, we have to find a method to achieve this. Here, we will attempt to employ three (3) methods.
- Data Rolling
- Exponential Decay Transformation
- Time Shift Transformation
- Logrithmic Transformation

### Data Rolling
**Change phrasing, this is copied from google** <br />
Rolling creates a rolling window with a specified size and perform calculations on the data in this window which, of course, rolls through the data. The figure below explains the concept of rolling. <br />
It is worth noting that the calculation starts when the whole window is in the data. In other words, if the size of the window is three, the first aggregation is done at the third.

```python
#Determine rolling statistics
rolmean = db.rolling(window=12).mean() #window size 12 denotes 12 months, giving rolling mean at yearly level
rolstd = db.rolling(window=12).std()
```

```python
#Plot rolling statistics
orig = plt.plot(db, color='blue', label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

The rolled data will observe a form as above.
### Data Made Stationary
#### Log Scale Transformation

```python
movingAverage = db_logScale.rolling(window=12).mean()
movingSTD = db_logScale.rolling(window=12).std()

db_rolling = db_logScale - movingAverage

#Remove NAN values
db_rolling.dropna(inplace=True)
```

#### Now to test if the new data is Stationary

```python
test_stationarity(db_rolling)
```

For this, we will selected the Critical value of 5% to represent $\alpha$ = -2.871388 <br />
The ADF Test Statistic observed is -3.296271 <br />
As ADF Test Statistic < $\alpha$, We reject $H_0$ and the data is stationary
### Exponential Decay Transformation

```python
db_exp_d = db_logScale.ewm(halflife=12, min_periods=0, adjust=True).mean()
db_exp_d = db_logScale - db_exp_d
test_stationarity(db_exp_d)
```

However, it appear this transformation was unable to make the data stationary and has an even bigger p-value than the previous transformation
### Time Shift Transformation

```python
db_TS = db_logScale - db_logScale.shift()
db_TS.dropna(inplace=True)
test_stationarity(db_TS)
```

Here, the Data is stationary. In addition, it appears the p-value is also much lower than the other 2 transformations. <br />
1) the p-value of 0.005197 is better than 0.015046 of the rolling mean <br />
2) the test statistic is very close to the Critical Value 
### So we will use the Time shift transformation on our data to make it stationary!
# (S)ARIMA Parameters
#### ACF and PACF plots

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
fig, ax = plt.subplots(2,1)
fig = sm.graphics.tsa.plot_acf(db_TS, lags=50, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(db_TS, lags=50, ax=ax[1])
plt.show()
```

https://www.datasciencecentral.com/profiles/blogs/tutorial-forecasting-with-seasonal-arima <br />
Identifying the order of differencing: <br />
d=0 if the series has no visible trend or ACF at all lags is low.<br />
d≥1 if the series has visible trend or positive ACF values out to a high number of lags.<br />
Note: if after applying differencing to the series and the ACF at lag 1 is -0.5 or more negative the series may be overdifferenced.<br />
Note: If you find the best d to be d=1 then the original series has a constant trend. A model with d=2 assumes that the original series has a time-varying trend.<br />
Identifying the number of AR and MA terms<br />
p is equal to the first lag where the PACF value is above the significance level.<br />
q is equal to the first lag where the ACF value is above the significance level.<br />
Identifying the seasonal part of the model:<br />
S is equal to the ACF lag with the highest value (typically at a high lag).<br />
D=1 if the series has a stable seasonal pattern over time.<br />
D=0 if the series has an unstable seasonal pattern over time.<br />
Rule of thumb: d+D≤2<br />
P≥1 if the ACF is positive at lag S, else P=0.<br />
Q≥1 if the ACF is negative at lag S, else Q=0.<br />
Rule of thumb: P+Q≤2

I personally prefer the much easier Auto_ARIMA which finds all the parameters of the model including the Seasonal parameters

```python
from pmdarima.arima import auto_arima
stepwise_model = auto_arima(db_TS, start_p=0, start_q=0,
                           max_p=4, max_q=4, 
                           start_P=0, 
                           d = 0 , D = 0, max_D = 1, m=12,
                            trace=True,seasonal=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(stepwise_model.aic())
```

So, the best set of paramters to select is  **ARIMA(0,0,0)(2,0,1)[12]** I selected m = 12 to represent the yearly number of months each 'season' runs for 

```python
model = sm.tsa.statespace.SARIMAX(db_TS, order=(0,0,0), seasonal_order = (2,0,1,12)).fit(max_iter=50, method='powell')
res = model.resid
fig, ax = plt.subplots(2,1)
fig = sm.graphics.tsa.plot_acf(res, lags=50, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(res, lags=50, ax=ax[1])
plt.show()
```

By Observation, after the implementation of the 'best' parameters, there are significantly fewer spikes in the ACF and PACF graphs

# Model Validations

```python
validation=y['2011-09':].index
start = y.index.get_loc('2011-09') 
pred1 = []
pred2 = []
pred3 = []
pred4 = []
pred5 = []
actual = []

logy = np.log(y)
for i in range(start, len(y)):
    
    actual.append(y.iloc[i]) 
    pred1.append(y.iloc[i-1]) 
    
    model2 = forecast.ses(y.iloc[:i]) 
    model2.fit()
    pred2.append(model2.forecast(1)[0]) 
    
    
    model3 = forecast.holtwinters(y.iloc[:i], additive = False,damped = False, m = 12)
    model3.fit()
    pred3.append(model3.forecast(1)[0]) 
    
    
    model4 = forecast.holtwinters(y.iloc[:i], additive = False,damped = True, m = 12)
    model4.fit()
    pred4.append(model4.forecast(1)[0]) 
    
    model5 = forecast.holtwinters(logy.iloc[:i],additive = True,damped = False, m = 12) 
    model5.fit()
    re5 = (logy.iloc[:i]-model5.smooth())[12:]
    pre = np.exp(model5.forecast(1)[0]) * np.mean(np.exp(re5))
    pred5.append(pre) 
    
    
    
columns=['Simple Random walk','Simple exponential smoothing', 
             'Holt-Winter Multiplicative','Holt-Winter Multiplicative Damping',"Holt Winters Log (Additive)",'Actual']
results = np.vstack([pred1,pred2,pred3,pred4,pred5, actual]).T
results = pd.DataFrame(results, columns=columns, index=validation)
```

```python
from statlearning import rmse_jack

table = pd.DataFrame(0.0, index=results.columns[:-1], columns=['RMSE','SE'])
for i in range(5):
    table.iloc[i,0], table.iloc[i,1] = rmse_jack(results.iloc[:,i], results.iloc[:,-1])
table.round(2)
```

### Conclusion

Overall, it appears that the Holt-Winters model performed better than the SARIMA model in terms of RMSE (SARIMA’s 28119.7431, Holt-Winters’ 21452.8870). <br />

The forecasts made by both models however are subject to the limitations of the data. As the data does not include many predictors, the training set does not have many points to teach the model about potential relationships. Additionally, the date column of the data did not have much date time information besides the month and the year. The day of the month had to be assumed as the first of each month. This meant that the models were unable to use weekends, holidays and other such key dates of the year to aid in the model forecasting process. Lastly, there was no meta data provided to know what country this data might have been collected from.<br />
If more date time values are added, meta data, predictors are available, we would be able to train a more powerful model to improve the forecasting power of the selected models. Lastly, although mentioned but not used, potentially resampling the data or shifting the timeseries data using lags could improve the forecasting prediction of the models.
